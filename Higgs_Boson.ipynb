```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # data visualization library  
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('training.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
list = ['EventId', 'Weight', 'Label']
X_train = train_df.drop(list, axis = 1)
X_train.head()
```

```python
Y_train = train_df[['Label']]
Y_train.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

le.fit(Y_train['Label'])
Y_train['Label'] = le.transform(Y_train['Label'])

Y_train.head()
```

```python
from sklearn.impute  import SimpleImputer

imputer = SimpleImputer(missing_values = -999.0, strategy = 'most_frequent')
X_train = imputer.fit_transform(X_train)
```

```python
X_train
```

```python
from sklearn.ensemble import GradientBoostingClassifier as GBC

gbc = GBC(n_estimators=50, max_depth=5,min_samples_leaf=200,max_features=10,verbose=1)
gbc.fit(X_train,Y_train) 
```

```python
train_predict = gbc.predict_proba(X_train)[:,1]

list = ['EventId']
X_test = test_df.drop(list, axis = 1)

Y_predict = gbc.predict_proba(X_test)[:,1]
Y_predict
```

```python
pcut = np.percentile(train_predict,85)
label_test = Y_predict > pcut

label_test
```

```python
result = []
for x in range(len(test_df['EventId'])):
    result.append([int(test_df['EventId'][x]), Y_predict[x], 's'*(label_test[x]==True)+'b'*(label_test[x]==False)])

result
```

```python
result = sorted(result, key = lambda x: x[1])

for i in range(len(result)):
    result[i][1] = i+1

result
```

```python
#sort according to the index
result = sorted(result, key = lambda x: x[0])
```

```python
file = open('submission.csv','w')
file.write('EventId,RankOrder,Class\n')

for entry in result:
    ent = str(entry[0]) + ',' + str(entry[1]) + ',' + str(entry[2]) + '\n'
    file.write(ent)
file.close()
```
